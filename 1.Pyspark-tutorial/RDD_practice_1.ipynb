```python
# sc = SparkContext('local', 'myfriend')

conf = SparkConf().setMaster("local").setAppName("myfriend")
sc = SparkContext.getOrCreate(conf=conf)

#sc.stop()
```

```python
sc
```

```python
lines = sc.textFile(path + '/fakefriends-noheader.csv')
```

```python
lines.collect() # id, , ,  
```

```python
#  line (('key'),  ('value')) 

rdd = lines.map(lambda line : (int(line.split(',')[2]), int(line.split(',')[3])))
rdd.collect()
```

```python
#   ( , 1) 

rdd = rdd.mapValues(lambda x : (x, 1))
rdd.collect()
```

```python
#  rdd   

totalByAge = rdd.reduceByKey(lambda x, y : (x[0] + y[0], x[1] + y[1]))
totalByAge.collect() #  : (, (   ,    ))
```

```python
averageByAge = totalByAge.mapValues(lambda x : x[0] / x[1])
# averageByAge.collect()
```

```python
results = averageByAge.sortBy(lambda x : x[0])
results.collect() # (,     )
```
