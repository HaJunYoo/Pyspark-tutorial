```python
import findspark
findspark.init()

from pyspark.sql import SparkSession

spark = SparkSession.builder.appName("spark3_test").master("local[*]").getOrCreate()
 
import pyspark
```

```python
# sc = SparkContext('local', 'myfriend')

conf = SparkConf().setMaster("local").setAppName("myfriend")
sc = SparkContext.getOrCreate(conf=conf)

#sc.stop()
```

```python
sc
```

```python
lines = sc.textFile(path + '/fakefriends-noheader.csv')
```

```python
lines.collect() # id, 이름, 나이, 친구 수
```

```python
# 각 line을 (나이('key'), 친구 수('value'))로 변환

rdd = lines.map(lambda line : (int(line.split(',')[2]), int(line.split(',')[3])))
rdd.collect()
```

```python
# 친구 수를 (친구 수, 1)로 변환

rdd = rdd.mapValues(lambda x : (x, 1))
rdd.collect()
```

```python
# 변환된 rdd를 나이별 합계 구하기

totalByAge = rdd.reduceByKey(lambda x, y : (x[0] + y[0], x[1] + y[1]))
totalByAge.collect() # 결과 : (나이, (해당 나이의 친구수의 총합, 해당 나이의 사람 수))
```

```python
averageByAge = totalByAge.mapValues(lambda x : x[0] / x[1])
# averageByAge.collect()
```

```python
results = averageByAge.sortBy(lambda x : x[0])
results.collect() # (나이, 해당 나이의 평균 친구 수)
```
