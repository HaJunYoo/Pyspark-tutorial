```python
from pyspark import SparkContext, SparkConf
```

```python
conf = SparkConf().setMaster("local").setAppName("wordcount")
sc = SparkContext.getOrCreate(conf=conf)
```

```python
sc
```

```python
path = '/content/drive/MyDrive/Spark/RDD_tutorial'
```

```python
book = sc.textFile(path+ '/book.txt')
```

```python
book.collect()
```

```python
#  ''     

book = book.filter(lambda x: len(x) > 0)
book.collect()
```

```python
#         
# ['The', 'Project', 'Gutenberg'..]

words = book.flatMap(lambda x: x.split())
words.collect()
```

```python
#     

words = words.map(lambda x: ''.join([str(i) for i in x if i.isalnum()]))
words.collect()
```

```python
#      ''  ( ''   )

words = words.filter(lambda x: len(x) > 0)
```

```python
#       

words = words.map(lambda x: ''.join([str(i.lower()) for i in x]))
words.collect()
```

```python
#  
# en_stops    
#pip install nltk
import nltk
from nltk.corpus import stopwords

nltk.download('stopwords')
en_stops = set(stopwords.words('english'))

words = words.filter(lambda x: x not in en_stops)
words.collect()
```

```python
# key-value rdd    
# ( ,  ) 

words = words.map(lambda x: (x, len(x)))
words.collect()
```

```python
# key  value        
# x+y =  *  

words = words.reduceByKey(lambda x,y : x+y)
words.collect()
```

```python
# value      

words = words.sortBy(lambda x: -x[1])
words.collect()
```
