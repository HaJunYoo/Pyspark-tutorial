```python
import findspark
findspark.init()

from pyspark.sql import SparkSession

spark = SparkSession.builder.appName("spark3_test").master("local[*]").getOrCreate()

import pyspark
```

```python
from pyspark import SparkContext, SparkConf
```

```python
conf = SparkConf().setMaster("local").setAppName("wordcount")
sc = SparkContext.getOrCreate(conf=conf)
```

```python
sc
```

```python
path = '/content/drive/MyDrive/Spark/RDD_tutorial'
```

```python
book = sc.textFile(path+ '/book.txt')
```

```python
book.collect()
```

```python
# 값이 ''인 레코드를 제거한 후 배열로 출력

book = book.filter(lambda x: len(x) > 0)
book.collect()
```

```python
# 각 레코드를 공백을 기준으로 분리 후 하나의 배열로 출력
# ['The', 'Project', 'Gutenberg'..]

words = book.flatMap(lambda x: x.split())
words.collect()
```

```python
# 각 레코드에 포함된 특수문자를 제거

words = words.map(lambda x: ''.join([str(i) for i in x if i.isalnum()]))
words.collect()
```

```python
# 특수문자만 있던 단어가 삭제되어 생긴 '' 값 제거(위에서 ''를 제거한 연산과 동일)

words = words.filter(lambda x: len(x) > 0)
```

```python
# 단어들을 모두 소문자로 변경한 후 배열로 출력

words = words.map(lambda x: ''.join([str(i.lower()) for i in x]))
words.collect()
```

```python
# 불용어 처리
# en_stops에 있는 단어만 배열로 출력
#pip install nltk
import nltk
from nltk.corpus import stopwords

nltk.download('stopwords')
en_stops = set(stopwords.words('english'))

words = words.filter(lambda x: x not in en_stops)
words.collect()
```

```python
# key-value rdd로 변환한 후 배열로 출력
# (각 단어, 단어 길이) 형식

words = words.map(lambda x: (x, len(x)))
words.collect()
```

```python
# key를 기준으로 value 값을 모두 더하는 연산을 한 뒤 배열로 출력
# x+y = 길이 * 개수 값

words = words.reduceByKey(lambda x,y : x+y)
words.collect()
```

```python
# value값을 기준으로 내림차순 정렬한 후 배열로 출력

words = words.sortBy(lambda x: -x[1])
words.collect()
```
